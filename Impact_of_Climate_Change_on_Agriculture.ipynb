
## <font color='green'> <b>Is there any increasing trend in the Extreme Weather Events?</b><font color='black'>   


```python
plt.figure(figsize=(12, 4)) 
barplot = sns.barplot(data=df2, x='Year', y='Extreme_Weather_Events')
plt.tight_layout
barplot.set_xticklabels(barplot.get_xticklabels(), rotation=45)
plt.show()
```

##### <font color='blue'> <b>According to this dataset we can conclude that there is no evidence for climate change because there is not any increasing trend of bad weather events</b><font color='black'>
